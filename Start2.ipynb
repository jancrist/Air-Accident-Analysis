```python
import pandas as pd
```

Creamos una funcion que remplaza los valores de columna

```python
def RemplazaValoresColumna(dataframe,columna_in_str,to_replace_in_str,replaced_by_in_str):
     dataframe[columna_in_str]=dataframe[columna_in_str].astype('string').str.replace(to_replace_in_str,replaced_by_in_str)
```

```python
RemplazaValoresColumna(adicional,'Incident_Date','00','1')
```

```python
adicional['Incident_Date'] = pd.to_datetime(adicional.Incident_Date, errors = 'coerce')

```
